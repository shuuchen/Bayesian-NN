```python
# https://docs.pymc.io/notebooks/sampler-stats.html
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import pymc3 as pm

%matplotlib inline
```

```python
with pm.Model() as model:
    mu1 = pm.Normal('mu1', mu=0, sd=1, shape=10)
```

```python
with model:
    step = pm.NUTS()
    trace = pm.sample(2000, tune=1000, init=None,
                     step=step, cores=2)
```

```python
trace.stat_names
```

```python
plt.plot(trace['step_size_bar'])
```

```python
sizes1, sizes2 = trace.get_sampler_stats('depth',
                                        combine=False)
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True,
                              sharey=True)
ax1.plot(sizes1)
ax2.plot(sizes2)
```

```python
accept = trace.get_sampler_stats('mean_tree_accept', burn=1000)
sb.distplot(accept, kde=True)
```

```python
accept.mean()
```

```python
trace['diverging'].nonzero()
```

```python
energy = trace['energy']
energy_diff = np.diff(energy)

sb.distplot(energy - energy.mean(), label='energy',
           kde=False)
sb.distplot(energy_diff, label='energy diff', kde=False)
plt.legend()
```

### Multiple samplers

```python
with pm.Model() as model:
    mu1 = pm.Bernoulli('mu1', p=0.8)
    mu2 = pm.Normal('mu2', mu=0, sd=1, shape=10)
```

```python
with model:
    step1 = pm.BinaryMetropolis([mu1])
    step2 = pm.Metropolis([mu2])
    trace = pm.sample(10000, init=None, step=[step1, step2],
                     cores=2, tune=1000)
```

```python
trace.stat_names
```

```python
pp=trace.get_sampler_stats('accept')
pp
```
